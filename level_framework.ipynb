```python
import pandas as pd
from pathlib import Path
import numpy as np

# Define path for saving
data_path = Path("data")
data_path.mkdir(exist_ok=True)

# Parameters
num_levels = 50
base_xp = 100

# Create DataFrame
levels = list(range(1, num_levels + 1))
base_xp = 100
growth_rate = 1.02  # 2% harder each level
xp_for_level = []

for level in levels:
    xp = base_xp * (growth_rate ** (level - 1))
    if level % 10 == 0:  # every 10th level, increase XP requirement
        xp *= 1.5
    xp_for_level.append(int(xp))

xp_total = np.cumsum(xp_for_level)


level_framework_df = pd.DataFrame({
    "level": levels,
    "xp_for_level": xp_for_level,
    "xp_total": xp_total
})

# Save to CSV
output_file = data_path / "level_framework_df.csv"
level_framework_df.to_csv(output_file, index=False)

print(f"✅ Saved level framework to {output_file}")
level_framework_df.head(10)
```
