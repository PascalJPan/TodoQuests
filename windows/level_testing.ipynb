```python
import math

def _xp_for_level(level: int, base: float, mult: float, exp: float) -> float:
    if level <= 0:
        return 0.0
    return base * level + mult * math.pow(level, exp)

def _xp_for_last_level(level: int, base: float, mult: float, exp: float) -> float:
    return _xp_for_level(level,base,mult,exp) - _xp_for_level(level-1,base,mult,exp)
```

```python
level_dif = 50
base_dif = 500
mult = 10 
exp = 2.2
```

```python
_xp_for_last_level(level_dif,base_dif,mult,exp)
```

```python
_xp_for_level(level_dif,base_dif,mult,exp)
```
